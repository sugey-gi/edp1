$$
1. \text{Calcular la Serie de Fourier en cosenos de } f(x) = e^{2x} \text{ para } 0 \leq x \leq 1.
$$


```python
from sympy import fourier_series, pi, plot, Piecewise, exp, cos
from sympy.abc import x

# Definir la función f(x) = e^(2x) en el intervalo [0, 1]
f = exp(2 * x)

# Calcular la Serie de Fourier en cosenos
s = fourier_series(f, (x, 0, 1))

# Truncar la serie para diferentes valores de n
s1 = s.truncate(n=3)  # Aproximación con n=3
s2 = s.truncate(n=5)  # Aproximación con n=5
s3 = s.truncate(n=7)  # Aproximación con n=7

# Graficar la función original y las aproximaciones de la serie
p = plot(f, s1, s2, s3, (x, 0, 1), show=False, legend=True)

# Etiquetas para la leyenda
p[0].label = 'f(x) = e^(2x)'
p[1].label = 'n=3'
p[2].label = 'n=5'
p[3].label = 'n=7'

# Mostrar la gráfica
p.show()
```

$$
2. \text{Calcular la Serie de Fourier en senos de } f(x) = e^{2x} \text{ para } 0 \leq x \leq 1.
$$


```python
from sympy import fourier_series, pi, plot, Piecewise, exp, sin
from sympy.abc import x

# Definir la función f(x) = e^(2x) en el intervalo [0, 1]
f = exp(2 * x)

# Extensión impar de f(x) en el intervalo [-1, 1]
f_impar = Piecewise(
    (f, x >= 0),
    (-f.subs(x, -x), x < 0)
)

# Calcular la Serie de Fourier en senos
s = fourier_series(f_impar, (x, -1, 1))

# Truncar la serie para diferentes valores de n
s1 = s.truncate(n=3)  # Aproximación con n=3
s2 = s.truncate(n=5)  # Aproximación con n=5
s3 = s.truncate(n=7)  # Aproximación con n=7

# Graficar la función original y las aproximaciones de la serie
p = plot(f, s1, s2, s3, (x, 0, 1), show=False, legend=True)

# Etiquetas para la leyenda
p[0].label = 'f(x) = e^(2x)'
p[1].label = 'n=3'
p[2].label = 'n=5'
p[3].label = 'n=7'

# Mostrar la gráfica
p.show()
```
