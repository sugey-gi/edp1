Problema de Ondas en una Cuerda

Grafique la función
$$ \( f(x') = \exp\left(-\frac{(x'-0.5)^2}{0.01}\right) \) $$
y suponga que esta representa una perturbación en una cuerda.

 Mostrar que al hacer el cambio de variable $$\( x' = x - ct \)$$ la perturbación se desplaza a la derecha.
    
Mostrar que al hacer el cambio de variable $$\( x' = x + ct \) $$ la perturbación se desplaza a la izquierda.
    
Graficar la superposición de las ondas viajeras de los incisos (a) y (b).


```python
#importamos la biblioteca requerida
import numpy as np
import matplotlib.pyplot as plt

# parametros de ocngiguracion
c = 1.0  # Velocidad de propagacion de la onda
x = np.linspace(0, 1, 500)  # Puntos en el espacio

# Definimos la función gaussiana
def f(x_prime):
    return np.exp(-((x_prime - 0.5)**2)/0.01)

# Creamos la figura
plt.figure(figsize=(10, 8))

# (a) Onda que viaja a la derecha (x - ct)
t_valores = [0, 0.2, 0.4]  # Tres tiempos diferentes
for t in t_valores:
    plt.plot(x, f(x - c*t), label=f't={t}')

plt.title('(a) Onda viajando a la derecha: f(x-ct)')
plt.xlabel('Posición x')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.show()

# (b) Onda que viaja a la izquierda (x + ct)
plt.figure(figsize=(10, 8))
for t in t_valores:
    plt.plot(x, f(x + c*t), label=f't={t}')

plt.title('(b) Onda viajando a la izquierda: f(x+ct)')
plt.xlabel('Posición x')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.show()

# (c) Superposición de ambas ondas
plt.figure(figsize=(10, 8))
plt.subplot(2, 1, 2)
for t in [0, 0.1, 0.2, 0.3]:
    y_superposition = f(x - c*t) + f(x + c*t)
    plt.plot(x, y_superposition, label=f't={t}')
plt.title('(c) Superposición de ondas viajeras\n$f(x-ct) + f(x+ct)$')
plt.xlabel('x')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
